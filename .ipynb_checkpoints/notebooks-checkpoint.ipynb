```python
# Load the top 10 players with the most goals
from db_views import get_top_10_players_with_most_goals



top10_players = get_top_10_players_with_most_goals("postgres")
```

```python
# use matplotlib to visualize the top 10 players with the most goals
import matplotlib.pyplot as plt

# Extract player names and expected goals from the list of dictionaries
player_names = [player['player_name'] for player in top10_players]
expected_goals = [player['expected_goals'] for player in top10_players]

# Create horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(player_names, expected_goals, color='cyan')

# Adding labels and title
plt.xlabel('Expected Goals')
plt.ylabel('Player Name')
plt.title('Top 10 Players with the Most Expected Goals')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the chart
plt.show()
```
